```python
import pandas as pd

df = pd.read_csv('../data/raw/thrift_data.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
# total revnue per row
df['revenue'] = df['final_price'] * df['quantity']

# total cost
df['total_cost'] = df['cost'] * df['quantity']

# total profit
df['total_profit'] = df['profit'] * df['quantity']
```

```python
# Core Business questions:
# 1) Which category makes the most profit?
df.groupby('category')['total_profit'].sum().sort_values(ascending=False)

# 2) Which items sell fastest?
df.groupby('item_name')['days_to_sell'].mean().sort_values()

# 3) Which source is most profitable?
df.groupby('source')['total_profit'].sum().sort_values(ascending=False)

# 4) Which brand performs best?
df.groupby('brand')['total_profit'].sum().sort_values(ascending=False)

# 5) What customer behavior is observed?
df.groupby('customer_type')['total_profit'].sum()
# Are returning customers worth more?

# 6) What is the customer payment behaviour?
df.groupby("payment_method")["total_profit"].sum()

df.head()
```
