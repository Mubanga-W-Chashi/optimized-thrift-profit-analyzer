```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
```

```python
items = ["Hoodie", "Jeans", "Sneakers", "T-Shirt", "Jacket"]
brands = ["Nike", "Adidas", "Zara", "H&M", "Puma"]
conditions = ["New", "Good", "Worn"]
sources = ["Market", "Supplier", "Donation"]
customer_types = ["New", "Returning"]
payment_methods = ["Cash", "Mobile Money"]
```

```python
data = []

for i in range(200):
    item = random.choice(items)
    brand = random.choice(brands)
    condition = random.choice(conditions)
    source = random.choice(sources)
    
    cost = round(random.uniform(5, 30), 2)
    
    markup = random.uniform(1.5, 3)
    selling_price = round(cost * markup, 2)
    
    discount = round(random.uniform(0, 5), 2)
    final_price = round(selling_price - discount, 2)
    
    profit = round(final_price - cost, 2)
    
    quantity = random.randint(1, 5)
    days_to_sell = random.randint(1, 30)
    
    date_sold = datetime.now() - timedelta(days=random.randint(0, 60))
    
    data.append([
        item, brand, condition, source,
        cost, selling_price, discount, final_price, profit,
        quantity, date_sold, days_to_sell,
        random.choice(customer_types),
        random.choice(payment_methods)
    ])
```
